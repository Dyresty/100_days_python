## Day 7 - Hangman


```python
import random
import time
from IPython.display import clear_output
from words import word_list 
from art import hangmen, banner

# Choose a random word
ch = random.randint(0, len(word_list) - 1)
word = word_list[ch].lower()
length = len(word)

# Initialize guessed word with underscores
guessed_word = ['_' if char != ' ' else ' ' for char in word]

# Display banner
print(banner)
time.sleep(2)

lives = 6
win = False

while lives > 0 and not win:
    print(' '.join(guessed_word))
    print(hangmen[lives])

    # Get user input
    guess = input("\nEnter a character to check\n").lower()
    clear_output(wait=True)
    
    print(f"You guessed {guess.upper()}")

    if guess in word:
        # Replace underscores with correctly guessed letter
        for i in range(len(word)):
            if word[i] == guess:
                guessed_word[i] = guess
        
        # Check for win condition
        if '_' not in guessed_word:
            win = True
            print(' '.join(guessed_word).upper())
    else:
        lives -= 1
        print(f"Lives remaining: {lives}")

# Print final result
clear_output(wait=True)
print(hangmen[lives])

if win:
    print(' '.join(guessed_word).upper())
    print("WIN")
else:
    print("LOSS")
    print(f"The word was {word.upper()}")
```

## Day 8 - Caeser Cipher

```python
def encrypt(n,str1,str2):
    length=len(str1)
    #A-Z => 65-90 , a-z => 97-122
    for i in range(0,length):
        if str1[i]>=chr(65) and str1[i]<=chr(90):
            if chr(ord(str1[i])+n)>chr(90):
                x=ord(str1[i])+n-90
                str2.append(chr(65+x-1))
            else:
                str2.append(chr(ord(str1[i])+n))
        else:
            str2.append(str1[i])

def decrypt(n,str1,str2):
    length=len(str1)
    for i in range(0,length):
        if str1[i]>=chr(65) and str1[i]<=chr(90):
            if chr(ord(str1[i])-n)<chr(65):
                x=ord(str1[i])+n-65
                str2.append(chr(90-x+1))
            else:
                str2.append(chr(ord(str1[i])-n))
        else:
            str2.append(str1[i])
                
ch=0
while ch==0:
    print("Welcome to Cipher\nEnter 1 for Encrypting\nElse proceeding for Decrypting\n")
    cho=int(input())
    if cho==1:
        text="encrypt"
    else: 
        text="decrypt"
    str1 = input(f"Enter the message to be {text}ed").upper()
    n = int(input("Enter the number"))
    n = n%26
    str2 = []
    if cho==1:
        encrypt(n,str1,str2)
    else:
        decrypt(n,str1,str2)
    print(str1)
    print("".join(str2))
    ch=int(input("Enter 0 to continue"))    
```

```python
MENU = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
            "milk":0
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}

resources = {
    "coffee":100,
    "water":300,
    "milk":200
}

money=0

sys_on=1

def suf_res(ch):
    z=1
    for i in resources:
        if resources[i]-MENU[ch]['ingredients'][i]<0:
            z=0
    if z==1:
        return 1
    return 0

def incr_bal(ch):
    global money
    money+=MENU[ch]['cost']

def decr_res(ch):
    for i in resources:
        resources[i]-=MENU[ch]['ingredients'][i]
    
while sys_on==1:
    choice=input("Enter choice:\nEspresso\nLatte\nCappuccino\n").lower()
    if choice=="report":
        print(f"\nREPORT\nWater = {resources['water']}\nMilk = {resources['milk']}\nCoffee = {resources['coffee']}\nMoney = {money}")
    elif choice=="off":
        print("\nPOWERING OFF\n")
        sys_on=0
    else:
        if choice=='espresso' or choice=='latte' or choice=='cappuccino':
            x=suf_res(choice)
            if x==1:
                currency=float(input(f"Available.\nTransfer currency of {MENU[choice]['cost']}\n"))
                if currency==MENU[choice]['cost']:
                    print("Thank you for buying coffee with us")
                    incr_bal(choice)
                    decr_res(choice)
                elif currency>MENU[choice]['cost']:
                    print(f"Returning balance of {float(round(currency-MENU[choice]['cost'],2))}")
                    print("Thank you for buying coffee with us")
                    incr_bal(choice)
                    decr_res(choice)
                else:
                    print(f"Returning currency of {float(round(currency,2))} as it is not sufficient")
            else:
                print("Resources not available. Sorry.")
        else:
            print("Invalid choice")
```
